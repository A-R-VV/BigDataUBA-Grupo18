```python
import wbdata
import pandas as pd
import matplotlib.pyplot as plt

# Definimos los países y sus indicadores
paises = ['ARG', 'MEX']
indicador = {"SP.POP.TOTL": "Población"}
fecha_inicio = 2000
fecha_fin = 2019

# Descargamos los datos a utilizar en el grafico
df = wbdata.get_dataframe(indicador, country=paises)
df.reset_index(inplace=True)
df["date"] = df["date"].astype(int)

# Filtramos por el rango de años
df = df[(df["date"] >= fecha_inicio) & (df["date"] <= fecha_fin)]


df_pivot = df.pivot(index="date", columns="country", values="Población")

# Convertimos a millones y redondeamos a 1 decimal
df_pivot = (df_pivot / 1_000_000).round(1)

# Gráfico estilo pyplot
plt.figure(figsize=(12, 7))
plt.plot(df_pivot.index, df_pivot["Argentina"], label="Argentina", color="blue", marker="o")
plt.plot(df_pivot.index, df_pivot["Mexico"], label="México", color="black", marker="o")

plt.title("Evolución de la Población Total (en millones): Argentina vs México")
plt.xlabel("Año")
plt.ylabel("Población (millones)")
plt.xticks(df_pivot.index, df_pivot.index.astype(int), rotation=45)  # Asegura que los años se muestren como enteros
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import wbdata
import pandas as pd
import matplotlib.pyplot as plt

# Configuramos los paises e indicadores
paises = ['ARG', 'MEX']
indicador = {"NY.GDP.PCAP.CD": "PIB per cápita"}
fecha_inicio = 2000
fecha_fin = 2023

# Descargamos datos
df = wbdata.get_dataframe(indicador, country=paises)
df.reset_index(inplace=True)
df["date"] = df["date"].astype(int)

# Filtrar por años deseados
df = df[(df["date"] >= fecha_inicio) & (df["date"] <= fecha_fin)]


df_pivot = df.pivot(index="date", columns="country", values="PIB per cápita")

# Redondear valores a enteros 
df_pivot = df_pivot.round(0)

# Crear figura y ejes (estilo orientado a objetos)
fig, ax = plt.subplots(figsize=(12, 7))

# Graficar líneas
ax.plot(df_pivot.index, df_pivot["Argentina"], label="Argentina", color="orange", marker="o")
ax.plot(df_pivot.index, df_pivot["Mexico"], label="México", color="green", marker="o")

# Ajustamos los titulos del grtafico
ax.set_title("Evolución del PIB per cápita (USD actuales): Argentina vs México")
ax.set_xlabel("Año")
ax.set_ylabel("PIB per cápita (USD)")

# Eje X con años redondeados
ax.set_xticks(df_pivot.index)
ax.set_xticklabels(df_pivot.index.astype(int), rotation=45)


ax.grid(True, linestyle='--', alpha=0.7)
ax.legend()


plt.tight_layout()
plt.show()
```

#### Ejercicio 4
De la página de noticias del [diario La Nación](https://www.lanacion.com.ar/) o cualquier diario que les interese, utilicen herramientas de web scraping para obtener los **links** de las noticias de la portada. Guarden los links obtenidos en un dataframe y expórtenlo a un archivo de excel.

Nota 1: es posible que logren obtener los links a las noticias sin el dominio: "https://www.lanacion.com.ar/". De ser así, concatenen el dominio a la ruta del link obtenido, tal que se obtenga un link al que se pueda acceder. Es decir, que las cadenas de caracteres finales tendrán la forma: https://www.lanacion.com.ar/*texto_obtenido*)

Nota 2: junto con su entrega, adjunten una captura de la página de noticias al momento de correr su código. Eso servirá al momento de la corrección para verificar que los links obtenidos hacen referencia a las noticias de ese día y hora.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de La Nación
url = "https://www.lanacion.com.ar/"

# Hacer la solicitud HTTP
headers = {"User-Agent": "Mozilla/5.0"}  # Para evitar bloqueos
response = requests.get(url, headers=headers)

# Verificar que la solicitud fue exitosa
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")

    # Extraer links y títulos de noticias
    noticias = []
    for article in soup.find_all("article"):  # Buscar dentro de <article>
        a_tag = article.find("a", href=True)
        if a_tag and "/nota/" in a_tag["href"]:
            full_link = "https://www.lanacion.com.ar" + a_tag["href"] if a_tag["href"].startswith("/") else a_tag["href"]
            titulo = article.find("h2") or article.find("h3")  # Buscar títulos dentro del article
            titulo_texto = titulo.text.strip() if titulo else "(Sin título)"
            noticias.append((titulo_texto, full_link))


    # Guardar en un DataFrame y exportar a Excel
    df = pd.DataFrame(noticias, columns=["Título", "Link"])
    df.to_excel("noticias_lanacion.xlsx", index=False)

  
```
