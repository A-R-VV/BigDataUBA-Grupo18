```python
import wbdata
import pandas as pd

# Configuramos los paises a trabajar y los indicadores elegidos
paises = ["AR", "MEX"]
indicadores = {"SP.POP.TOTL": "Población total", "NY.GDP.PCAP.CD": "PIB per cápita (USD actuales)"}

# Obtenemos los datos del Banco Mundial
datos = wbdata.get_dataframe(indicadores, country=paises)

# Separamos los datos legidos
datos_argentina = datos.loc[datos.index.get_level_values('country') == 'Argentina']
datos_mexico = datos.loc[datos.index.get_level_values('country') == 'Mexico']

# Mostramos los resultados 
print("Datos de Argentina:")
print(datos_argentina)

print("\nDatos de México:")
print(datos_mexico)
```

```python
import wbdata
import pandas as pd
import matplotlib.pyplot as plt

# Definimos los países y sus indicadores
paises = ['ARG', 'MEX']
indicador = {"SP.POP.TOTL": "Población"}
fecha_inicio = 2000
fecha_fin = 2019

# Descargamos los datos a utilizar en el grafico
df = wbdata.get_dataframe(indicador, country=paises)
df.reset_index(inplace=True)
df["date"] = df["date"].astype(int)

# Filtramos por el rango de años
df = df[(df["date"] >= fecha_inicio) & (df["date"] <= fecha_fin)]


df_pivot = df.pivot(index="date", columns="country", values="Población")

# Convertimos a millones y redondeamos a 1 decimal
df_pivot = (df_pivot / 1_000_000).round(1)

# Gráfico estilo pyplot
plt.figure(figsize=(12, 7))
plt.plot(df_pivot.index, df_pivot["Argentina"], label="Argentina", color="blue", marker="o")
plt.plot(df_pivot.index, df_pivot["Mexico"], label="México", color="black", marker="o")

plt.title("Evolución de la Población Total (en millones): Argentina vs México")
plt.xlabel("Año")
plt.ylabel("Población (millones)")
plt.xticks(df_pivot.index, df_pivot.index.astype(int), rotation=45)  # Asegura que los años se muestren como enteros
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import wbdata
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Configuramos los paises y sus indicadores
paises = ['ARG', 'MEX']
indicador = {"NY.GDP.PCAP.CD": "PIB per cápita"}
fecha_inicio = 2000
fecha_fin = 2023

# Descargamos y preparamos los datos
df = wbdata.get_dataframe(indicador, country=paises)
df.reset_index(inplace=True)
df["date"] = df["date"].astype(int)
df = df[(df["date"] >= fecha_inicio) & (df["date"] <= fecha_fin)]
df_pivot = df.pivot(index="date", columns="country", values="PIB per cápita")
df_pivot = df_pivot.reindex(pd.Index(range(fecha_inicio, fecha_fin + 1), name="date"))
df_pivot = df_pivot.round(0)

# Gráfico de barras agrupadas (orientado a objetos)
fig, ax = plt.subplots(figsize=(14, 7))

anios = df_pivot.index
x = np.arange(len(anios))
width = 0.35

ax.bar(x - width/2, df_pivot['Argentina'], width, label='Argentina', color='grey')
ax.bar(x + width/2, df_pivot['Mexico'], width, label='México', color='red')

# Personalizamos el grafico agusto
ax.set_title("PIB per cápita (USD actuales): Argentina vs México (2000-2023)", fontsize=14)
ax.set_xlabel("Año", fontsize=12)
ax.set_ylabel("PIB per cápita (USD)", fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(anios, rotation=45)
ax.legend()
ax.grid(True, linestyle='--', axis='y', alpha=0.7)

plt.tight_layout()
plt.show()  
```

#### Ejercicio 4
De la página de noticias del [diario La Nación](https://www.lanacion.com.ar/) o cualquier diario que les interese, utilicen herramientas de web scraping para obtener los **links** de las noticias de la portada. Guarden los links obtenidos en un dataframe y expórtenlo a un archivo de excel.

Nota 1: es posible que logren obtener los links a las noticias sin el dominio: "https://www.lanacion.com.ar/". De ser así, concatenen el dominio a la ruta del link obtenido, tal que se obtenga un link al que se pueda acceder. Es decir, que las cadenas de caracteres finales tendrán la forma: https://www.lanacion.com.ar/*texto_obtenido*)

Nota 2: junto con su entrega, adjunten una captura de la página de noticias al momento de correr su código. Eso servirá al momento de la corrección para verificar que los links obtenidos hacen referencia a las noticias de ese día y hora.

```python
import requests  # Importa la librería para hacer solicitudes HTTP
from bs4 import BeautifulSoup  # Importa BeautifulSoup para analizar HTML
import pandas as pd  # Importa pandas para trabajar con DataFrames
from IPython.display import display, HTML  # Importa funciones para mostrar HTML en Jupyter

# Configuración inicial
url = "https://www.lanacion.com.ar/"  # URL del sitio web a extraer
headers = {"User-Agent": "Mozilla/5.0"}  # Encabezados para simular un navegador

try:
    # 1. Obtener el HTML de la página
    response = requests.get(url, headers=headers, timeout=10)  # Realiza la solicitud HTTP
    response.raise_for_status()  # Lanza una excepción si la solicitud falla
    soup = BeautifulSoup(response.text, 'html.parser')  # Analiza el HTML con BeautifulSoup

    # 2. Extraer los contenedores principales de las noticias
    news_containers = soup.find_all(['article', 'section'], class_=lambda x: x and (
        'modulo' in x or 'com-card' in x or 'description-container' in x
    ))  # Busca los contenedores de noticias

    # 3. Filtrar y guardar los links de las noticias
    news_data = []  # Lista para almacenar los datos de las noticias
    for container in news_containers:  # Itera sobre los contenedores de noticias
        link_tag = container.find_parent('a', href=True) or container.find('a', href=True)  # Busca la etiqueta <a>
        
        if link_tag:  # Si se encontró una etiqueta <a>
            href = link_tag['href']  # Obtiene el atributo href
            full_url = f"https://www.lanacion.com.ar{href}" if href.startswith('/') else href  # Construye la URL completa
            
            title = container.find(['h1', 'h2', 'h3'], class_=lambda x: x and 'title' in x)  # Busca el título
            title_text = title.get_text(strip=True) if title else "Sin título"  # Obtiene el texto del título

            news_data.append({"Título": title_text, "Link": full_url})  # Agrega los datos a la lista

    # 4. Eliminar duplicados y mostrar los primeros 10 resultados (lista ampliada)
    df = pd.DataFrame(news_data).drop_duplicates(subset=['Link']).head(10).copy()  # Crea un DataFrame y elimina duplicados

    # Función para crear el HTML del enlace
    def crear_html_link(row):
        link = row['Link']  # Obtiene el enlace
        titulo = row['Título']  # Obtiene el título
        return f'<a href="{link}" title="{titulo}" target="_blank">{titulo[:30]}...</a>'  # Crea el HTML del enlace

    # Aplicar la función para crear la columna HTML del link
    df['Link_HTML'] = df.apply(crear_html_link, axis=1)  # Aplica la función a cada fila del DataFrame

    if not df.empty:  # Si se encontraron noticias
        # Construir el HTML personalizado para mostrar los enlaces debajo de los títulos
        html_output = "<table><tr><th>Título</th></tr>"  # Inicia la tabla HTML
        for index, row in df.iterrows():  # Itera sobre las filas del DataFrame
            html_output += f"<tr><td>{row['Título']}</td></tr>"  # Agrega el título a la tabla
            html_output += f"<tr><td>{row['Link_HTML']}</td></tr>"  # Agrega el enlace HTML a la tabla
        html_output += "</table>"  # Cierra la tabla HTML
        
        # Mostrar el HTML personalizado
        display(HTML(html_output))  # Muestra el HTML en Jupyter
        
        df.to_excel('portada_la_nacion_links.xlsx', index=False)  # Guarda el DataFrame en un archivo Excel
        print(f"Se guardaron {len(df)} noticias principales.")  # Imprime un mensaje de éxito
    else:
        print("No se encontraron noticias con la estructura esperada.")  # Imprime un mensaje de error

except requests.exceptions.RequestException as e:  # Captura errores de solicitudes HTTP
    print(f"Error al acceder a la página: {e}")  # Imprime el error
except Exception as e:  # Captura otros errores
    print(f"Error: {e}")  # Imprime el error
```
