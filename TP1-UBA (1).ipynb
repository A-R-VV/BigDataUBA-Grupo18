#### Ejercicio 3 - Practicando con Matplotlib
Armen dos gráficos distintos usando la librería Matplotlib (repasen Clase 4). Uno programandolo con el estilo *pyplot* y otro gráfico de estilo *orientada a objetos*

```python
# Resolver acá estilo pyplot
```

```python
import wbdata
import pandas as pd
import matplotlib.pyplot as plt

# Paso 1: Configuración de países e indicadores
paises = ["AR", "MEX"]
indicadores = {"SP.POP.TOTL": "Población total", "NY.GDP.PCAP.CD": "PIB per cápita (USD actuales)"}

# Paso 2: Obtención de datos del Banco Mundial
datos = wbdata.get_dataframe(indicadores, country=paises)

# Paso 3: Separar datos por país (ajustado para el índice)
datos_argentina = datos.loc[datos.index.get_level_values('country') == 'Argentina']
datos_mexico = datos.loc[datos.index.get_level_values('country') == 'Mexico']

# Paso 4: Convertir el índice de fechas a datetime
fechas_argentina = pd.to_datetime(datos_argentina.index.get_level_values('date'))
fechas_mexico = pd.to_datetime(datos_mexico.index.get_level_values('date'))

# Paso 5: Gráfico con estilo pyplot (corregido)
plt.figure(figsize=(10, 6))  # Tamaño del gráfico
plt.plot(fechas_argentina, datos_argentina[('PIB per cápita (USD actuales)')], label='Argentina')
plt.plot(fechas_mexico, datos_mexico[('PIB per cápita (USD actuales)')], label='México')
plt.title('PIB per cápita en Argentina y México')
plt.xlabel('Año')
plt.ylabel('PIB per cápita (USD)')
plt.legend()
plt.show()

# Paso 6: Gráfico con estilo orientado a objetos (corregido)
fig, ax = plt.subplots(figsize=(10, 6))  # Tamaño del gráfico

ultimo_anio_argentina = datos_argentina[('Población total')].iloc[-1]
ultimo_anio_mexico = datos_mexico[('Población total')].iloc[-1]

ax.bar(['Argentina', 'México'], [ultimo_anio_argentina, ultimo_anio_mexico])
ax.set_title('Población total en Argentina y México (último año)')
ax.set_ylabel('Población')
plt.show()
```
